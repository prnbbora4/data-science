```
import pandas as pd
```

```
import matplotlib.pyplot as plt
```

```
a=[1,2,3]
b=[4,5,6]
```

```
plt.plot(a,b)
plt.title("Graph")
plt.xlabel("a")
plt.ylabel("b")
#plt.legend([a,b])
plt.show()
```

```
c=[7,8,9]
```

```
plt.plot(a,b, color="r")
plt.plot(a,c)
plt.title("Graph")
plt.xlabel("a")
plt.ylabel("b")
plt.legend(["ab","ac"])
plt.show()
```

```
plt.bar(a,b,color="black")
plt.bar(b,c, color="g")
plt.plot()
plt.title("Graph")
plt.xlabel("a")
plt.ylabel("b")
plt.show()
```

```
plt.bar(a,b, alpha=1, color="black")
plt.bar(a,c, alpha=0.5, color="g")
plt.plot()
plt.title("Graph")
plt.xlabel("a")
plt.ylabel("b")
plt.show()
```

```
plt.scatter(a,b)
plt.scatter(b, c, marker='v', color='r')
plt.scatter(c, a, marker='^', color='m')
plt.show()
```

```
plt.stackplot(a,b)
plt.stackplot(b, c, color='r')
plt.stackplot(c, a, color='m')
plt.show()
```

```
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(a,b,c, c='b', marker='o', label='blue')
ax.scatter(c,b,a, c='g', marker='D', label='green')
plt.tight_layout()
plt.show()
```
